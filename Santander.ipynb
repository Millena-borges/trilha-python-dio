```python
import json
```

```python
users = [
    {'id': 1, 'name': 'Millena', 'news': []},
    {'id': 2, 'name': 'Maria', 'news': []},
    {'id': 3, 'name': 'Matheus', 'news': []}
]

print(f"Extraídos {len(users)} usuários.")
```

Transform

```python
!pip install openai
```

```python
def generate_ai_news(user):
    # Aqui entraria a chamada da API da OpenAI
    return f"Olá {user['name']},Você é um especialista em markting bancário. Invista agora e garanta seu futuro ninja!"

for user in users:
  news = generate_ai_news(user)
  print(news)
```

LOAD (Carga)

```python
print("\nRelatório Final:")
for user in users:
    print(f"Usuário: {user['name']} | Mensagem: {user['news'][0]['description']}")

# Opcional: Salvar em um arquivo físico para finalizar o projeto
with open('SDW_Resultados.json', 'w', encoding='utf-8') as f:
    json.dump(users, f, indent=2, ensure_ascii=False)

print("\nProcesso ETL concluído! Dados salvos com suscesso.")
```
